```python
import numpy as np  # Importa o pacote numÃ©rico.
import matplotlib.pyplot as plt  # Importa o pacote de plotagem (pyplot->submodulo do matplotlib)
import pandas as pd  # Importa pandas.

%matplotlib inline
```

```python
def make_index(df):
    return [int(idx.split('.')[1]) for idx
            in df['Sample']]
```

```python
kw = dict(skiprows=1, na_values='NaN', sep=',',
          skipinitialspace=True, index_col=False)

df = pd.read_table('BTS_AllCores.csv', **kw)
df.index = make_index(df)
```

```python
Descritive_CO_CO1 = np.nanmean(df['CO1_%C']), np.nanmedian(df['CO1_%C']), np.nanmin(df['CO1_%C']), np.nanmax(df['CO1_%C'])
Descritive_NT_CO1 = np.nanmean(df['CO1_%N']), np.nanmedian(df['CO1_%N']), np.nanmin(df['CO1_%N']), np.nanmax(df['CO1_%N'])
Descritive_dC_CO1 = np.nanmean(df['CO1_d13C']), np.nanmedian(df['CO1_d13C']), np.nanmin(df['CO1_d13C']), np.nanmax(df['CO1_d13C'])
Descritive_dN_CO1 = np.nanmean(df['CO1_d15N']), np.nanmedian(df['CO1_d15N']), np.nanmin(df['CO1_d15N']), np.nanmax(df['CO1_d15N'])
Descritive_CN_CO1 = np.nanmean(df['CO1_C/N']), np.nanmedian(df['CO1_C/N']), np.nanmin(df['CO1_C/N']), np.nanmax(df['CO1_C/N'])
print('Core CO1')
print('COT', Descritive_CO_CO1, 'TN', Descritive_NT_CO1, 'd13C', Descritive_dC_CO1, 'd13N', Descritive_dN_CO1, 'CN', Descritive_CN_CO1)

Descritive_CO_CO2 = np.nanmean(df['CO2_%C']), np.nanmedian(df['CO2_%C']), np.nanmin(df['CO2_%C']), np.nanmax(df['CO2_%C'])
Descritive_NT_CO2 = np.nanmean(df['CO2_%N']), np.nanmedian(df['CO2_%N']), np.nanmin(df['CO2_%N']), np.nanmax(df['CO2_%N'])
Descritive_dC_CO2 = np.nanmean(df['CO2_d13C']), np.nanmedian(df['CO2_d13C']), np.nanmin(df['CO2_d13C']), np.nanmax(df['CO2_d13C'])
Descritive_dN_CO2 = np.nanmean(df['CO2_d15N']), np.nanmedian(df['CO2_d15N']), np.nanmin(df['CO2_d15N']), np.nanmax(df['CO2_d15N'])
Descritive_CN_CO2 = np.nanmean(df['CO2_C/N']), np.nanmedian(df['CO2_C/N']), np.nanmin(df['CO2_C/N']), np.nanmax(df['CO2_C/N'])
print('Core CO2')
print('COT', Descritive_CO_CO2, 'TN', Descritive_NT_CO2, 'd13C', Descritive_dC_CO2, 'd13N', Descritive_dN_CO2,'CN', Descritive_CN_CO2)

Descritive_CO_CO4 = np.nanmean(df['CO4_%C']), np.nanmedian(df['CO4_%C']), np.nanmin(df['CO4_%C']), np.nanmax(df['CO4_%C'])
Descritive_NT_CO4 = np.nanmean(df['CO4_%N']), np.nanmedian(df['CO4_%N']), np.nanmin(df['CO4_%N']), np.nanmax(df['CO4_%N'])
Descritive_dC_CO4 = np.nanmean(df['CO4_d13C']), np.nanmedian(df['CO4_d13C']), np.nanmin(df['CO4_d13C']), np.nanmax(df['CO4_d13C'])
Descritive_dN_CO4 = np.nanmean(df['CO4_d15N']), np.nanmedian(df['CO4_d15N']), np.nanmin(df['CO4_d15N']), np.nanmax(df['CO4_d15N'])
Descritive_CN_CO4 = np.nanmean(df['CO4_C/N']), np.nanmedian(df['CO4_C/N']), np.nanmin(df['CO4_C/N']), np.nanmax(df['CO4_C/N'])
print('Core CO4')
print('COT', Descritive_CO_CO4, 'TN', Descritive_NT_CO4, 'd13C', Descritive_dC_CO4, 'd13N', Descritive_dN_CO4,'CN', Descritive_CN_CO4)

Descritive_CO_CO5 = np.nanmean(df['CO5_%C']), np.nanmedian(df['CO5_%C']), np.nanmin(df['CO5_%C']), np.nanmax(df['CO5_%C'])
Descritive_NT_CO5 = np.nanmean(df['CO5_%N']), np.nanmedian(df['CO5_%N']), np.nanmin(df['CO5_%N']), np.nanmax(df['CO5_%N'])
Descritive_dC_CO5 = np.nanmean(df['CO5_d13C']), np.nanmedian(df['CO5_d13C']), np.nanmin(df['CO5_d13C']), np.nanmax(df['CO5_d13C'])
Descritive_dN_CO5 = np.nanmean(df['CO5_d15N']), np.nanmedian(df['CO5_d15N']), np.nanmin(df['CO5_d15N']), np.nanmax(df['CO5_d15N'])
Descritive_CN_CO5 = np.nanmean(df['CO5_C/N']), np.nanmedian(df['CO5_C/N']), np.nanmin(df['CO5_C/N']), np.nanmax(df['CO5_C/N'])
print('Core CO5')
print('COT', Descritive_CO_CO5, 'TN', Descritive_NT_CO5, 'd13C', Descritive_dC_CO5, 'd13N', Descritive_dN_CO5,'CN', Descritive_CN_CO5)


```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(ncols=2, nrows=2, figsize=(8,8))

kw = dict(linestyle='none', markersize=8,
          alpha=0.95, markeredgecolor= 'w')

#Plotagem
#CxN
ax0.plot(df['CO1_%C'], df['CO1_%N'], marker='d', markerfacecolor='SaddleBrown',**kw)
ax0.plot(df['CO2_%C'], df['CO2_%N'], marker='o', markerfacecolor='MediumSeaGreen',**kw)
ax0.plot(df['CO4_%C'], df['CO4_%N'], marker='^', markerfacecolor='OrangeRed',**kw)
ax0.plot(df['CO5_%C'], df['CO5_%N'], marker='p', markerfacecolor='DarkSlateBlue',**kw)

#C/N x d13C
ax1.plot(df['CO1_C/N'], df['CO1_d13C'], marker='d', markerfacecolor='SaddleBrown',**kw)
ax1.plot(df['CO2_C/N'], df['CO2_d13C'], marker='o', markerfacecolor='MediumSeaGreen',**kw)
ax1.plot(df['CO4_C/N'], df['CO4_d13C'], marker='^', markerfacecolor='OrangeRed',**kw)
ax1.plot(df['CO5_C/N'], df['CO5_d13C'], marker='p', markerfacecolor='DarkSlateBlue',**kw)

#d13C x d15N
ax2.plot(df['CO1_d13C'], df['CO1_d15N'], marker='d', markerfacecolor='SaddleBrown',**kw)
ax2.plot(df['CO2_d13C'], df['CO2_d15N'], marker='o', markerfacecolor='MediumSeaGreen',**kw)
ax2.plot(df['CO4_d13C'], df['CO4_d15N'], marker='^', markerfacecolor='OrangeRed',**kw)
ax2.plot(df['CO5_d13C'], df['CO5_d15N'], marker='p', markerfacecolor='DarkSlateBlue',**kw)

#C/N x d15N
ax3.plot(df['CO1_C/N'], df['CO1_d15N'], marker='d', markerfacecolor='SaddleBrown',**kw)
ax3.plot(df['CO2_C/N'], df['CO2_d15N'], marker='o', markerfacecolor='MediumSeaGreen',**kw)
ax3.plot(df['CO4_C/N'], df['CO4_d15N'], marker='^', markerfacecolor='OrangeRed',**kw)
ax3.plot(df['CO5_C/N'], df['CO5_d15N'], marker='p', markerfacecolor='DarkSlateBlue',**kw)

#Nomes dos eixos X
ax0.set_xlabel(r'C$_{\mathbf{org}}$ (%)')
ax1.set_xlabel(r'C$_{\mathbf{org}}$/TN')
ax2.set_xlabel(r'$\delta^{13}$C ($\perthousand$)')
ax3.set_xlabel(r'C$_{\mathbf{org}}$/TN')

#Nomes dos eixos Y
ax0.set_ylabel(r'Total N (%)')
ax1.set_ylabel(r'$\delta^{13}$C ($\perthousand$)')
ax2.set_ylabel(r'$\delta^{15}$N ($\perthousand$)')
ax3.set_ylabel(r'$\delta^{15}$N ($\perthousand$)')

#Intervalos dos eixos X e Y
#Intervalos dos eixos X e Y
ax0.axis([0, 4, 0.1, 0.3])
ax1.axis([0, 18, -27,-19])
ax2.axis([-27,-19, 3, 7])
ax3.axis([0, 18, 3, 7])

#PosiÃ§Ã£o do nome/ticks dos eixos (x e y)
ax0.xaxis.set_label_position('top')
ax0.xaxis.set_ticks_position('top')

ax1.xaxis.set_label_position('top')
ax1.xaxis.set_ticks_position('top')
ax1.yaxis.set_label_position('right')
ax1.yaxis.set_ticks_position('right')

ax3.yaxis.set_label_position('right')
ax3.yaxis.set_ticks_position('right')

#NÃºmero de ticks (x e y)
ax0.locator_params(axis='x', tight=True, nbins=5)
ax1.locator_params(axis='x', tight=True, nbins=5)
ax1.locator_params(axis='y', tight=True, nbins=6)
ax2.locator_params(axis='x', tight=True, nbins=6)
ax2.locator_params(axis='y', tight=True, nbins=6)
ax3.locator_params(axis='x', tight=True, nbins=5)
ax3.locator_params(axis='y', tight=True, nbins=5)

#EspaÃ§amento entre plots
fig.subplots_adjust(wspace=0.05, hspace=0.05)

#Mostrar gride
ax0.grid(True)
ax1.grid(True)
ax2.grid(True)
ax3.grid(True)
```
