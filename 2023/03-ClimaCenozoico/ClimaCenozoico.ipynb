### Artigo Base
Westerhold(2020) - An astronomically dated record of Earth’s climate and its predictability over the last 66 million years117110, Science (369), 6509 DOI:10.1126/science.aba6853

```python
import pandas as pd

df = pd.read_csv("TableS34.tab", skiprows=22, sep="\t")
df.head()
```

```python
import matplotlib.pyplot as plt

fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(11, 6), sharex=True)

df.plot(
    ax=ax0,
    x="Tuned time [Ma]",
    y="Foram benth δ18O [‰ PDB] (ISOBENd18oLOESSsmooth)",
    legend=False,
)

df.plot(
    ax=ax0,
    x="Tuned time [Ma]",
    y="Foram benth δ18O [‰ PDB] (ISOBENd18oLOESSsmoothLongTerm)",
    legend=False,
    color="red",
)


df.plot(
    ax=ax1,
    x="Tuned time [Ma]",
    y="Foram benth δ13C [‰ PDB] (ISOBENd13cLOESSsmooth)",
    legend=False,
)

df.plot(
    ax=ax1,
    x="Tuned time [Ma]",
    y="Foram benth δ13C [‰ PDB] (ISOBENd13cLOESSsmoothLongTerm)",
    legend=False,
    color="red",
)


ax0.set_ylabel("benthic δ$^{18}$O (‰)")
ax1.set_ylabel("benthic δ$^{13}$C (‰)")

ax0.invert_yaxis()
ax0.invert_xaxis()
ax0.grid()
ax1.grid()

ax0.set_xlim(68, 0);
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(10, 5), sharey=True)

cs = ax0.scatter(
    df["Foram benth δ13C [‰ PDB] (ISOBENd13cLOESSsmooth)"],
    df["Foram benth δ18O [‰ PDB] (ISOBENd18oLOESSsmooth)"],
    s=10,
    c=df["Tuned time [Ma]"],
)

ax0.set_xlabel("benthic δ$^{13}$C (‰)")
ax0.set_ylabel("benthic δ$^{18}$O (‰)")


cs = ax1.scatter(
    df["Foram benth δ13C [‰ PDB] (ISOBENd13cLOESSsmoothLongTerm)"],
    df["Foram benth δ18O [‰ PDB] (ISOBENd18oLOESSsmoothLongTerm)"],
    s=10,
    c=df["Tuned time [Ma]"],
)

ax1.set_xlabel("benthic δ$^{13}$C (‰)")
ax1.set_ylabel("benthic δ$^{18}$O (‰)")


fig.colorbar(cs)
```
