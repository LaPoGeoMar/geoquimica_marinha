## Ciclo do N

Ver figura: https://www.nature.com/articles/ismej20078/figures/1

```python
import pandas as pd

df_bulk = pd.read_csv(
    "ODP_Site-964-967_TOC_TN.tab",
    skiprows=39,
    sep="\t",
)

df_bulk.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15, 3))

df_bulk['TOC [%]'].plot(ax=ax, kind="line");
```

```python
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(5, sharex=True)
fig.suptitle('COT, NT, d13C, d15N e COT/NT')

ax0.plot(df_bulk.index, df_bulk['TOC [%]'])
ax1.plot(df_bulk.index, df_bulk['TN [%]'])  
ax2.plot(df_bulk.index, df_bulk['δ13C TOC [‰ PDB]'])
ax3.plot(df_bulk.index, df_bulk['δ15N [‰ air]'])
ax4.plot(df_bulk.index, df_bulk['TOC/TN'])
```

```python
df_biom = pd.read_csv(
    "ODP_Site-964-967_biomarker-abundance.tab",
    skiprows=52,
    sep="\t",
)

df_biom.head()
```

```python
fig, (ax0, ax1, ax2, ax3) = plt.subplots(4, sharex=True)
fig.suptitle('COT, δ15N, fBHT-II, fHG-C5')

ax0.plot(df_bulk.index, df_bulk['TOC [%]'])
ax1.plot(df_bulk.index, df_bulk['δ15N [‰ air]'])
ax2.plot(df_biom.index, df_biom['fBHT-II'])
ax3.plot(df_biom.index, df_biom['fHG-C5'])
```
