```python
#qual a biblioteca que preciso importar para análise e estrutura de dados?
import pandas as pd

#criar o dataframe para ler a terceira aba da tabela Aula01_BCF.xlsx

#df = pd.read_excel("Aula01_BCF.xlsx", sheet_name=2)

#########################################################################

#reparar nos "nd" presentes nas colunas
#substituir com NaN
#SHIF+TAB para procurar na_values

df = pd.read_excel("Aula01_BCF.xlsx", sheet_name=2, na_values="nd")

df
```

```python
#Transformar pg/L em ng/L

df["Água(ngL-1)"] = df["Água(pg/L)"] / 1000
df
```

```python
#Calcular BCF
import numpy as np

df["logBCF"] = np.log10(df["Mexilhão(ng/g-lw)"] / df["Água(ngL-1)"])

df
```

```python
#gráfico de espalhamento 
%matplotlib inline

df.plot.scatter("logKow", "logBCF")
```
